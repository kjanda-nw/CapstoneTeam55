# Explore Forest Tracker Data Labels

```python
#Read in needed libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read in the train label data
target = pd.read_csv("../Data/train_v2.csv")
target.columns = ["image_name", "labels"]
target.head()
```

```python
#let's get a list of all the labels
diff_labels = []
for l in target['labels'].values:
    tags = l.split(' ')
    for label in tags:
        if label not in diff_labels:
            diff_labels.append(label)
            
#look at what we have
diff_labels
```

```python
#create encoding for the labels
for l in diff_labels:
    target[l] = target['labels'].apply(lambda x: 1 if l in x.split(' ') else 0)
    
    
target.head()
```

```python
#get label counts
target[diff_labels].sum()
```

```python
#cooccurence matrix
def make_cooccurance_matrix(label_category):
    numeric_df = target[label_category]; 
    c_matrix = numeric_df.T.dot(numeric_df)
    sns.heatmap(c_matrix)
    return c_matrix

plt.figure(figsize=(7,5))
plt.title('Co-Occurance Heatmap', fontsize=20)
make_cooccurance_matrix(diff_labels)
```

```python
weather_labels = ['clear', 'partly_cloudy', 'haze', 'cloudy']
make_cooccurance_matrix(weather_labels)
```

```python
land_labels = ['primary', 'agriculture', 'water', 'cultivation', 'habitation']
make_cooccurence_matrix(land_labels)
```

```python
rare_labels = [l for l in diff_labels if target[diff_labels].sum()[l] < 2000]
make_cooccurence_matrix(rare_labels)
```
